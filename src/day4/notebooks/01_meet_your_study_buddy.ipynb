## ğŸ­ **Experience 1: Meet Your TutorAgent**

Let's start by talking to your TutorAgent. This agent specializes in educational explanations and has access to:
- ğŸ” Web search for finding learning resources
- ğŸ’» Code analysis and examples
- ğŸ§  Memory of your learning progress
- ğŸ’ Emotional intelligence to adapt explanations

**Try asking about a topic you're learning or want to learn!**

```python
# ğŸ“š Chat with Your TutorAgent
# Ask about any topic you want to learn!

def chat_with_tutor(message):
    """Chat with the TutorAgent and see memory + emotional intelligence in action"""
    
    print(f"ğŸ‘¤ You: {message}")
    print("\nğŸ¤” TutorAgent is thinking...")
    
    # Analyze emotions first
    emotion_analysis = emotional_intelligence.analyze_emotion(message)
    print(f"ğŸ’ Emotion detected: {emotion_analysis['primary_emotion']} (confidence: {emotion_analysis['confidence_level']})")
    
    # Get response from TutorAgent (uses memory and emotional context)
    response = tutor_agent.teach(message, topic="general", understanding_level=5.0)
    
    # Store in memory - pass the full emotion analysis object
    conversation_memory.store_conversation(
        agent_type="tutor", 
        user_message=message, 
        agent_response=response,
        topic="general",
        emotion_data=emotion_analysis  # Pass the full emotion analysis
    )
    
    print(f"\nğŸ“š TutorAgent: {response}")
    print("\n" + "="*60)
    return response

# Example interactions - try these or ask your own questions!
print("ğŸ“ Let's have a conversation with your TutorAgent!")
print("Try asking about:") 
print("- 'Can you explain Python functions?'")
print("- 'I'm confused about machine learning'")
print("- 'Help me understand loops in programming'")
print("\nOr ask about any topic you're interested in!")
print("\n" + "="*60)

# Let's start with a friendly introduction
chat_with_tutor("Hi! I'm new to programming and want to learn Python. Can you help me?")
```

```python
# Ask a follow-up question to see memory in action!
chat_with_tutor("That's helpful! Can you show me a simple Python example to get started?")
```

```python
# Try expressing confusion to see emotional adaptation
chat_with_tutor("I'm still confused about how functions work. This is frustrating!")
```

## ğŸ­ **Experience 2: Your SessionAgent in Action**

Now let's meet your SessionAgent - your productivity and time management coach. This agent:
- â° Helps optimize your study sessions
- ğŸ“Š Tracks your focus and energy levels
- ğŸŒŸ Suggests breaks and productivity techniques
- ğŸ§  Remembers your study patterns and preferences

```python
# â° Chat with Your SessionAgent
# Ask about study planning, breaks, or productivity!

def chat_with_session_agent(message):
    """Chat with SessionAgent about study management"""
    
    print(f"ğŸ‘¤ You: {message}")
    print("\nğŸ¤” SessionAgent is analyzing your study needs...")
    
    # Analyze emotions and context
    emotion_analysis = emotional_intelligence.analyze_emotion(message)
    print(f"âš¡ Session context: {emotion_analysis.get('learning_state', emotion_analysis['primary_emotion'])} state detected")
    
    # Get response from SessionAgent (correct method name)
    response = session_agent.manage_time(message, current_topic="study_planning", available_time=60)
    
    # Store in memory
    conversation_memory.store_conversation(
        agent_type="session", 
        user_message=message, 
        agent_response=response,
        topic="study_planning",
        emotion_data=emotion_analysis
    )
    
    print(f"\nâ° SessionAgent: {response}")
    print("\n" + "="*60)
    return response

print("â° Let's talk to your SessionAgent about study management!")
print("Try asking about:")
print("- 'How should I plan my study session?'")
print("- 'I've been studying for 2 hours and feel tired'")
print("- 'What's the best way to focus while learning?'")
print("\n" + "="*60)

# Start with session planning
chat_with_session_agent("I want to start learning Python seriously. How should I structure my study sessions?")
```

```python
# Simulate being tired to see adaptive responses
chat_with_session_agent("I've been studying for 90 minutes and I'm feeling overwhelmed. Should I take a break?")
```

## ğŸ­ **Experience 3: Your GoalAgent - Achievement Tracker**

Finally, let's meet your GoalAgent - your personal achievement tracker and motivational coach. This agent:
- ğŸ¯ Helps you set and track learning goals
- ğŸ“ˆ Monitors your progress over time
- ğŸ‰ Celebrates your achievements
- ğŸ’ª Provides motivation during challenging times

```python
# ğŸ¯ Chat with Your GoalAgent
# Talk about learning goals, progress, and achievements!

def chat_with_goal_agent(message):
    """Chat with GoalAgent about goals and progress"""
    
    print(f"ğŸ‘¤ You: {message}")
    print("\nğŸ¤” GoalAgent is analyzing your progress...")
    
    # Analyze emotions and motivation level
    emotion_analysis = emotional_intelligence.analyze_emotion(message)
    print(f"ğŸ¯ Motivation level: {emotion_analysis['confidence_level']}")
    
    # Get response from GoalAgent (correct method name)
    response = goal_agent.coach(message, current_goal=None, progress_update=None)
    
    # Store in memory - pass the full emotion analysis object
    conversation_memory.store_conversation(
        agent_type="goal", 
        user_message=message, 
        agent_response=response,
        topic="goal_setting",
        emotion_data=emotion_analysis  # Pass the full emotion analysis
    )
    
    print(f"\nğŸ¯ GoalAgent: {response}")
    print("\n" + "="*60)
    return response

print("ğŸ¯ Let's talk to your GoalAgent about your learning journey!")
print("Try sharing:")
print("- 'I want to become a Python developer in 6 months'")
print("- 'I just understood how functions work!'")
print("- 'I'm struggling with this concept and feel discouraged'")
print("\n" + "="*60)

# Start with goal setting
chat_with_goal_agent("I want to learn Python programming and eventually build my own AI projects. Can you help me set some goals?")
```

```python
# Share an achievement to see celebration response
chat_with_goal_agent("Great news! I just wrote my first Python function and it worked perfectly! I'm so excited!")
```
